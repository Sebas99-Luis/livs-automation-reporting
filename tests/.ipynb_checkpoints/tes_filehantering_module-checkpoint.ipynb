```python
import sys
import os
from dotenv import load_dotenv

load_dotenv()

sys.path.append(r"C:\Users\info\Documents\NORRDATA\Proyectos\livs-automation-reporting\src")

from ingest.detect_files import get_incoming_files
from ingest.read_csv import read_csv_file
from clean.clean_record import clean_dataframe
from file_handling.move_files import move_file, write_log

# 1. Detectar archivo
files = get_incoming_files()
filepath = files[0]
filename = os.path.basename(filepath)

# 2. Leer archivo (solo para obtener filas)
df, missing = read_csv_file(filepath)

# 3. Forzar un error artificial
try:
    raise Exception("Simulated critical error for testing")

except Exception as e:
    target = move_file(filepath, "bad")
    write_log(filename, len(df), 0, str(e))
    print("Archivo movido a BAD:", target)
    print("Error registrado:", e)
```
